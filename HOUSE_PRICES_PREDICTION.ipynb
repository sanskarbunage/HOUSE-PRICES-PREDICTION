**SPLIT DATA**

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
for col in x.columns:
  plt.scatter(x[col],y)
  plt.xlabel(col)
  plt.ylabel('Price')
  plt.show()
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

**LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
ypred=model.predict(x_test)
```

```python
ypred
```

```python
a=model.coef_
b=model.intercept_
```

```python
print(a)
```

```python
print(b)
```

```python
model.predict(x[0:5])
```

```python
model.predict(x)
```

**MODEL EVALUATION**

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,ypred)
rmse=np.sqrt(mse)
print('RMSE value : {:.4f}'.format(rmse))
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,ypred)
print('r2 value : {:.4f}'.format(r2))
```

```python
# Predictions
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# R2 scores
print("Train R²:", r2_score(y_train, y_train_pred))
print("Test R²:", r2_score(y_test, y_test_pred))
```

**RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest=RandomForestRegressor(random_state=42)
forest.fit(x_train,y_train)
yf_pred=forest.predict(x_test)
```

```python
x_test
```

```python
yf_pred
```

**MODEL EVALUATION**

```python
msef=mean_squared_error(y_test,yf_pred)
frmse=np.sqrt(msef)
print('RMSE value : {:.4f}'.format(frmse))
```

```python
r2f=r2_score(y_test,yf_pred)
print('r2 value : {:.4f}'.format(r2f))
```

```python
# Predictions
ry_train_pred = forest.predict(x_train)
ry_test_pred = forest.predict(x_test)

# R2 scores
print("Train R²:", r2_score(y_train,ry_train_pred ))
print("Test R²:", r2_score(y_test, ry_test_pred))
```

**RANDOMIZED SEARCH CV**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {'n_estimators': [100, 200, 300],
              'max_depth': [None, 10, 20, 30],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 4]}
```

```python
random_search = RandomizedSearchCV(estimator=forest,param_distributions=param_dist,n_iter=10,cv=3,
                                   scoring='r2',n_jobs=-1,random_state=42)
random_search.fit(x_train,y_train)
```

```python
print('best parameter : ' ,random_search.best_params_)
best_model=random_search.best_estimator_
```

```python
rypred=best_model.predict(x_test)
```

**MODEL EVALUATION**

```python
R_mse=mean_squared_error(y_test,rypred)
R_rmse=np.sqrt(R_mse)
print('RMSE value : {:.4f}'.format(R_rmse))
```

```python
Rr2=r2_score(y_test,rypred)
print('r2 value : {:.4f}'.format(Rr2))
```

```python
# Predictions
Ry_train_pred = random_search.predict(x_train)
Ry_test_pred = random_search.predict(x_test)

# R2 scores
print("Train R²:", r2_score(y_train,Ry_train_pred ))
print("Test R²:", r2_score(y_test, Ry_test_pred))
```
